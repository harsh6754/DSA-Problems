```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\agarw\\Downloads\\homeprices.csv")
df
```

```python
plt.scatter(df.area,df.price,color='red',marker='*')
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
reg = linear_model.LinearRegression()
```

```python
X=df[['area']]
y=df['price']
```

```python
reg.fit(X,y)
```

```python
plt.scatter(df.area,df.price,color='red',marker='*')
plt.plot(df.area,reg.predict(X),color='blue',linewidth=2)
plt.show()
```
